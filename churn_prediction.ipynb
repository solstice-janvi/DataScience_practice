```python
import pandas as pd
```

```python
df = pd.read_csv('Customer-Churn.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols_to_encode = ['customerID','gender','Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','MonthlyCharges','TotalCharges','Churn']

for col in cols_to_encode:
    df[col] = le.fit_transform(df[col])
```

```python
inputs = df.drop('Churn',axis='columns')
inputs.head(10)
```

```python
target = df['Churn']
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```

```python
model.predict([[3962,1,0,0,0,34,1,0,0,2,0,2,0,0,0,1,0,3,498,1466]])
```
